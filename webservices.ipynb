```python
import requests

import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
import io
import base64
```

```python
result = requests.get('http://cat-fact.herokuapp.com/facts/random')
print(result)
print(result.json())
print(result.json()['text'])
```

```python
result = requests.get('http://3.22.117.102:5000/?msg=HelloWorld!')
print(result.json())
```

```python
result = requests.get('http://3.22.117.102:5000/',
                     params={'msg': 'Hello from params'})
print(result.json())
```

```python
result = requests.post('http://3.22.117.102:5000/',
                       json={'msg': 'Hello from data'})
print(result.json())
```

```python
image = open("luna.png", "rb").read()
encoded = base64.b64encode(image)

result = requests.get('http://3.22.117.102:5000/',
                     params={'msg': encoded})
encoded = result.json()['response']
imgData = base64.b64decode(encoded)
plt.imshow(np.array(Image.open(io.BytesIO(imgData))))
```
